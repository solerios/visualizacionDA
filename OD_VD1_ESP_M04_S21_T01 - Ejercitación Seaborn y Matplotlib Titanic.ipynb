```python
# Importar las librerías necesarias
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv('train_clean.csv')
```

```python
df
```

#### Features Description 

* PassengerId: Identificador único para cada pasajero.
* Survived: Indicador de si el pasajero sobrevivió (1) o no (0).
* Pclass: Clase del billete del pasajero (1, 2 o 3), representando la clase de viaje.
* Name: Nombre completo del pasajero.
* Sex: Género del pasajero (masculino o femenino).
* Age: Edad del pasajero en años.
* SibSp: Número de hermanos/cónyuges a bordo con el pasajero.
* Parch: Número de padres/hijos a bordo con el pasajero.
* Ticket: Número del billete.
* Fare: Tarifa pagada por el billete.
* Cabin: Número de cabina asignada al pasajero (puede ser NaN si no se asignó).
* Embarked: Puerto de embarque (C = Cherbourg, Q = Queenstown, S = Southampton).
* Family_Size: Tamaño de la familia

```python
fig = plt.figure(figsize=(20, 6))
gs = gridspec.GridSpec(nrows=1, ncols=3, figure=fig)

ax1 =fig.add_subplot(gs[0, 0])
sns.distplot(df['Age'])

ax2 =fig.add_subplot(gs[0, 1])
sns.boxplot(x='Pclass', y='Age', hue='Sex', data=df)

ax3 =fig.add_subplot(gs[0, 2])
sns.swarmplot(x='Pclass', y='Age', hue='Survived', data=df, dodge=True)
plt.legend(title='Sobrevivió?', labels=['No', 'Si'])
```

```python
sns.violinplot(x='Pclass', y='Age', hue='Survived', data=df, palette='viridis')
plt.title('Distribución de edad por clase y supervivencia')
plt.xlabel('Clase')
plt.ylabel('Edad')
plt.show()
```

Las personas que más sobrevivieron pertenecian a primera clase, tenían menos edad y eran mayormente mujeres.
